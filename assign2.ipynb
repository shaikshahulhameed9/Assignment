```python
# 1
import random
def coin_toss_simulation(trials=10000):
    heads = 0
    tails = 0
    
    for _ in range(trials):
        toss = random.choice(["H", "T"])  
        if toss == "H":
            heads += 1
        else:
            tails += 1

    prob_heads = heads / trials
    prob_tails = tails / trials
    
    print("Coin Toss Simulation (10,000 trials)")
    print(f"Heads: {heads}, Tails: {tails}")
    print(f"Probability of Heads ≈ {prob_heads:.4f}")
    print(f"Probability of Tails ≈ {prob_tails:.4f}\n")


def dice_roll_simulation(trials=10000):
    sum_seven = 0
    
    for _ in range(trials):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        if die1 + die2 == 7:
            sum_seven += 1

    prob_sum_seven = sum_seven / trials
    
    print("Two Dice Roll Simulation (10,000 trials)")
    print(f"Number of times sum = 7: {sum_seven}")
    print(f"Probability of sum = 7 ≈ {prob_sum_seven:.4f}\n")

coin_toss_simulation(10000)
dice_roll_simulation(10000)
```

```python
# 2
import random

def prob_at_least_one_six(trials=100000, rolls=10):
    success_count = 0
    
    for _ in range(trials):
        got_six = False
        for _ in range(rolls):
            if random.randint(1, 6) == 6:
                got_six = True
                break  
        if got_six:
            success_count += 1
    
    probability = success_count / trials
    return probability

estimated_prob = prob_at_least_one_six(100000, 10)
print(f"Estimated Probability of at least one '6' in 10 rolls ≈ {estimated_prob:.4f}")
```

```python
# 3
import random

def simulate_with_replacement(trials=1000):
    colors = ["Red"] * 5 + ["Green"] * 7 + ["Blue"] * 8

    draws = [random.choice(colors) for _ in range(trials)]

    prev_blue_count = 0        
    red_given_prev_blue = 0    

    prev_red_count = 0          
    blue_given_prev_red = 0     
    for i in range(1, len(draws)):
        prev = draws[i-1]
        curr = draws[i]

        if prev == "Blue":
            prev_blue_count += 1
            if curr == "Red":
                red_given_prev_blue += 1

        if prev == "Red":
            prev_red_count += 1
            if curr == "Blue":
                blue_given_prev_red += 1

    p_red_given_blue_emp = (red_given_prev_blue / prev_blue_count) if prev_blue_count else 0.0
    p_blue_given_red_emp = (blue_given_prev_red / prev_red_count) if prev_red_count else 0.0

    total_red = draws.count("Red")
    total_blue = draws.count("Blue")
    p_red_emp = total_red / trials
    p_blue_emp = total_blue / trials

    bayes_rhs_emp = (p_blue_given_red_emp * p_red_emp) / p_blue_emp if p_blue_emp else 0.0

    p_red_theo = 5 / 20
    p_blue_theo = 8 / 20
    p_blue_given_red_theo = p_blue_theo 
    bayes_rhs_theo = (p_blue_given_red_theo * p_red_theo) / p_blue_theo if p_blue_theo else 0.0
    p_red_given_blue_theo = p_red_theo  
    print(f"Simulation trials = {trials}")
    print("---- Empirical (simulation) results ----")
    print(f"P(Red | Previous = Blue)  ≈ {p_red_given_blue_emp:.4f}  (count {red_given_prev_blue}/{prev_blue_count})")
    print(f"P(Blue | Previous = Red)  ≈ {p_blue_given_red_emp:.4f}  (count {blue_given_prev_red}/{prev_red_count})")
    print(f"P(Red) empirical             = {p_red_emp:.4f}  (count {total_red}/{trials})")
    print(f"P(Blue) empirical            = {p_blue_emp:.4f}  (count {total_blue}/{trials})")
    print(f"Bayes RHS (empirical values) = {bayes_rhs_emp:.4f}")

    print("\n---- Theoretical results (with replacement -> independence) ----")
    print(f"P(Red) theoretical          = {p_red_theo:.4f}")
    print(f"P(Blue) theoretical         = {p_blue_theo:.4f}")
    print(f"P(Blue | Red) theoretical   = {p_blue_given_red_theo:.4f}")
    print(f"P(Red | Blue) theoretical   = {p_red_given_blue_theo:.4f}")
    print(f"Bayes RHS (theoretical)     = {bayes_rhs_theo:.4f}")

    return {
        "empirical": {
            "p_red_given_blue": p_red_given_blue_emp,
            "p_blue_given_red": p_blue_given_red_emp,
            "p_red": p_red_emp,
            "p_blue": p_blue_emp,
            "bayes_rhs": bayes_rhs_emp
        },
        "theoretical": {
            "p_red_given_blue": p_red_given_blue_theo,
            "p_blue_given_red": p_blue_given_red_theo,
            "p_red": p_red_theo,
            "p_blue": p_blue_theo,
            "bayes_rhs": bayes_rhs_theo
        }
    }

if __name__ == "__main__":
    simulate_with_replacement(trials=1000)
```

```python
# 4
import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.40]

sample = np.random.choice(values, size=1000, p=probabilities)

emp_mean = np.mean(sample)
emp_var = np.var(sample)      
emp_std = np.std(sample)      
print("Sample size: 1000")
print(f"Empirical Mean             = {emp_mean:.4f}")
print(f"Empirical Variance         = {emp_var:.4f}")
print(f"Empirical Standard Deviation = {emp_std:.4f}")
```

```python
# 5
import numpy as np
mean = 5
sample_size = 2000
samples = np.random.exponential(scale=mean, size=sample_size)

emp_mean = np.mean(samples)
emp_var = np.var(samples)
emp_std = np.std(samples)

theo_mean = mean
theo_var = mean**2
theo_std = mean

print("Exponential Distribution Simulation (mean=5, n=2000)")
print(f"Empirical Mean  = {emp_mean:.4f} | Theoretical Mean  = {theo_mean}")
print(f"Empirical Var   = {emp_var:.4f} | Theoretical Var   = {theo_var}")
print(f"Empirical Std   = {emp_std:.4f} | Theoretical Std   = {theo_std}")
```

```python
# 6
import numpy as np
uniform_data = np.random.uniform(0, 1, 10000)

n = 30
num_samples = 1000
sample_means = [np.mean(np.random.uniform(0, 1, n)) for _ in range(num_samples)]
sample_means = np.array(sample_means)

print("Uniform(0,1) distribution:")
print(f"Mean ≈ {np.mean(uniform_data):.4f}, Variance ≈ {np.var(uniform_data):.4f}")

print("\nDistribution of Sample Means (n=30):")
print(f"Mean ≈ {np.mean(sample_means):.4f}, Variance ≈ {np.var(sample_means):.4f}")
print(f"Std Dev ≈ {np.std(sample_means):.4f}")

theo_mean = 0.5
theo_var = 1/12   
theo_var_sample_mean = theo_var / n
print("\nTheoretical Expectations:")
print(f"E[X] = {theo_mean}, Var(X) = {theo_var:.4f}")
print(f"E[Sample Mean] = {theo_mean}, Var(Sample Mean] = {theo_var_sample_mean:.4f}")
```
